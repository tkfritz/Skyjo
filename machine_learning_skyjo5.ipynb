The different computer levels against each other. Testing the nwest level 5, first against the random level 0. 

```python
#5 against 1 takes forever 

names=('alpha','beta')
nature=('computer','computer')
levels=(6,2)
start_time=time.time()
nn=int(1000)
all_scores=np.zeros((7,nn))
for i in range(nn):
    print(f"running case {i}")
    scores,turns,last_player,output=skyjo_round(names,nature,levels,0,True,True,True)
    all_scores[0,i]=i
    all_scores[1:3,i]=scores
    all_scores[3,i]=turns
    all_scores[4,i]=np.argmin(scores)
    all_scores[5,i]=output[0,0]
    all_scores[6,i]=output[0,output.shape[1]-1]    
stop_time=time.time()
print(f"running time was {stop_time-start_time}")
print(all_scores.T)
np.savetxt("level6_level2_1000_1.txt",all_scores)
```

```python
all_scores=np.loadtxt("level5_level0_1000_0.txt")
deltalevel5=0
deltalevel0=0
for i in range(all_scores.shape[1]):
    if all_scores[1,i]<all_scores[2,i]:
        #to calculate by how much a  human wins in average 
        deltalevel5+=(all_scores[2,i]-all_scores[1,i])
    else:
        deltalevel0+=(all_scores[1,i]-all_scores[2,i])
print("level 5 gainst level 0")
print(f"level 5 won {int(1000-sum(all_scores[4]))} times of {all_scores.shape[1]} rounds")
print(f"the level 5 average is {round(np.mean(all_scores[1]),3)} points")
print(f"the level 0 average is {round(np.mean(all_scores[2]),3)} points")
print(f"the level 5 average adventage is  {round(np.mean(all_scores[2]-all_scores[1]),3)} points")
print(f"level 5 wins to {round(np.mean(100-all_scores[4]*100),1)} %")
print(f"level 5 average adventage {round(deltalevel5/(1000-sum(all_scores[4])),3)} points")
print(f"level 0 average adventage {round(deltalevel0/(sum(all_scores[4])),3)} points")
```

Level 5 performs very well against level 0. That is expected since it is trained on it. Thus it mainly tested that there was no coding mistake.
