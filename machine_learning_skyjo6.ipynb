The different computer levels against each other. Testing the newest level 9 and then 11, first against the random level 0. 

Then again level 5 which seems to be the best previous level (or level 1 but both are rather similar anyway). Since level 5 gets sometime sin an infinite loope, level 6 is used instead. 

```python
 

names=('alpha','beta')
nature=('computer','computer')
levels=(11,6)
start_time=time.time()
nn=int(500)
all_scores=np.zeros((7,nn))
for i in range(nn):
    print(f"running case {i}")
    scores,turns,last_player,output=skyjo_round(names,nature,levels,0,True,True,True)
    all_scores[0,i]=i
    all_scores[1:3,i]=scores
    all_scores[3,i]=turns
    all_scores[4,i]=np.argmin(scores)
    all_scores[5,i]=output[0,0]
    all_scores[6,i]=output[0,output.shape[1]-1]    
stop_time=time.time()
print(f"running time was {stop_time-start_time}")
print(all_scores.T)
np.savetxt("level11_level6_500_0.txt",all_scores)
```

```python
all_scores=np.loadtxt("level9_level0_500_0.txt")
deltalevel9=0
deltalevel0=0
for i in range(all_scores.shape[1]):
    if all_scores[1,i]<all_scores[2,i]:
        #to calculate by how much a  human wins in average 
        deltalevel9+=(all_scores[2,i]-all_scores[1,i])
    else:
        deltalevel0+=(all_scores[1,i]-all_scores[2,i])
print("level 9 gainst level 0")
print(f"level 9 won {int(all_scores.shape[1]-sum(all_scores[4]))} times of {all_scores.shape[1]} rounds")
print(f"the level 9 average is {round(np.mean(all_scores[1]),3)} points")
print(f"the level 0 average is {round(np.mean(all_scores[2]),3)} points")
print(f"the level 9 average adventage is  {round(np.mean(all_scores[2]-all_scores[1]),3)} points")
print(f"level 9 wins to {round(np.mean(100-all_scores[4]*100),1)} %")
print(f"level 9 average adventage {round(deltalevel9/(all_scores.shape[1]-sum(all_scores[4])),3)} points")
print(f"level 0 average adventage {round(deltalevel0/(sum(all_scores[4])),3)} points")


all_scores=np.loadtxt("level11_level0_500_0.txt")
deltalevel9=0
deltalevel0=0
for i in range(all_scores.shape[1]):
    if all_scores[1,i]<all_scores[2,i]:
        #to calculate by how much a  human wins in average 
        deltalevel9+=(all_scores[2,i]-all_scores[1,i])
    else:
        deltalevel0+=(all_scores[1,i]-all_scores[2,i])
print("level 11 gainst level 0")
print(f"level 11 won {int(all_scores.shape[1]-sum(all_scores[4]))} times of {all_scores.shape[1]} rounds")
print(f"the level 11 average is {round(np.mean(all_scores[1]),3)} points")
print(f"the level 0 average is {round(np.mean(all_scores[2]),3)} points")
print(f"the level 11 average adventage is  {round(np.mean(all_scores[2]-all_scores[1]),3)} points")
print(f"level 11 wins to {round(np.mean(100-all_scores[4]*100),1)} %")
print(f"level 11 average adventage {round(deltalevel9/(all_scores.shape[1]-sum(all_scores[4])),3)} is points")
print(f"level 0 average adventage {round(deltalevel0/(sum(all_scores[4])),3)} is points")
```
