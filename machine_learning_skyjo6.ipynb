```python
 

names=('alpha','beta')
nature=('computer','computer')
levels=(17,6)
start_time=time.time()
nn=int(100)
all_scores=np.zeros((7,nn))
for i in range(nn):
    print(f"running case {i}")
    scores,turns,last_player,output=skyjo_round(names,nature,levels,0,True,True,True)
    all_scores[0,i]=i
    all_scores[1:3,i]=scores
    all_scores[3,i]=turns
    all_scores[4,i]=np.argmin(scores)
    all_scores[5,i]=output[0,0]
    all_scores[6,i]=output[0,output.shape[1]-1]    
stop_time=time.time()
print(f"running time was {stop_time-start_time}")
print(all_scores.T)
np.savetxt("level17_level6_100_0.txt",all_scores)
#none type errror cannot get numeric of none type object problem 
```

```python
def combine_results(list_inputs,level0=0,level1=2,silent=False,losses=True,rounds=True):
    cc=0
    for i in range(len(list_inputs)):
        a=np.loadtxt(list_inputs[i])
        cc+=a.shape[1]
    res=np.zeros((a.shape[0],int(cc))) 
    cc=0
    for i in range(len(list_inputs)):
        a=np.loadtxt(list_inputs[i])
        res[:,cc:cc+a.shape[1]]=a
        cc+=a.shape[1]    
    deltalevela=0
    deltalevelb=0
    for i in range(res.shape[1]):
        if res[1,i]<res[2,i]:
            #to calculate by how first and second plkayer win in average
            deltalevela+=(res[2,i]-res[1,i])
        else:
            deltalevelb+=(res[1,i]-res[2,i])
    if rounds==True:
        print(f"level {int(level0)} against level {int(level1)} in rounds")
        
    print(f"to {np.round(np.mean(res[6])*100,3)} % finishes level {int(level1)}")    
    if losses==True:
        print(f"level {int(level0)} lost {(res.shape[1]-int(res.shape[1]-sum(res[4])))} times of {res.shape[1]} rounds")
    else:
        print(f"level {int(level1)} won {int(res.shape[1]-sum(res[4]))} times of {all_scores.shape[1]} rounds")
    print(f"the level {int(level0)} average is {round(np.mean(res[1]),3)} points")
    print(f"the level {int(level1)} average is {round(np.mean(res[2]),3)} points")
    print(f"the level {int(level0)} average adventage is  {round(np.mean(res[2]-res[1]),3)} points")
    print(f"level {int(level0)} wins to {round(np.mean(100-res[4]*100),1)} %")
    print(f"level {int(level1)} average win adventage is {round(deltalevela/(res.shape[1]-sum(res[4])),3)} points")
    print(f"level {int(level0)} average win adventage is {round(deltalevelb/(sum(res[4])),3)} points") 
    print(" ")
    return res
```

```python

level9level0=combine_results(["level9_level0_500_0.txt","level9_level0_1500_0.txt","level9_level0_3000_0.txt"],level0=9,level1=0)

level11level0=combine_results(["level11_level0_500_0.txt","level11_level0_1500_0.txt","level11_level0_3000_0.txt"],level0=11,level1=0)

level13level0=combine_results(["level13_level0_1000_0.txt","level13_level0_2500_0.txt","level13_level0_2000_0.txt"],level0=13,level1=0)

level15level0=combine_results(["level15_level0_1000_0.txt","level15_level0_2500_0.txt","level15_level0_2000_0.txt"],level0=15,level1=0)


level17level0=combine_results(["level17_level0_100_0.txt"],level0=17,level1=0)
```

Level 9 and level 11 performs very well against level 0. That is good and was not guranteed since it was trained mostly on rather few human data sets.

Level 13 (human) and 15 (win) perform similar as their sibling above, thus there was no necessary an improvement compatred to 9 and 11 as was the aim at least for 15. 13 might be worse but that is human imitation.
